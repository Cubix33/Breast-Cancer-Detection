```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_excel("data.xlsx")
```

# feature selection

```python
from sklearn.model_selection import train_test_split
```

```python
diag_map={'M':1, 'B':0}
```

```python
df['diagnosis']=df['diagnosis'].map(diag_map)
```

```python
df
```

```python
x=df[['radius_mean', 'perimeter_mean', 'area_mean', 'concavity_mean', 'concave points_mean']]
```

```python
y=df[['diagnosis']]
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test=train_test_split(df, y, test_size=0.2, random_state=42)
```

# model KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier()
```

```python
knn.fit(x_train, y_train)
```

```python
knn_y_pred=knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(knn_y_pred, y_test)
```

# Model-logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(random_state=0)
```

```python
lr.fit(x_train, y_train)
```

```python
lr_y_pred=lr.predict(x_test)
```

```python
accuracy_score(lr_y_pred, y_test)
```

# Model- Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(x_train, y_train)
```

```python
gnb_y_pred=gnb.predict(x_test)
```

```python
accuracy_score(gnb_y_pred, y_test)
```

# Model- K Cross Validation

```python
from sklearn.model_selection import cross_val_score
```

```python
accuracy_all=[]
cvs_all=[]
```

```python
scores=cross_val_score(knn, x, y, cv=10) #KNN
accuracy_all.append(accuracy_score(knn_y_pred, y_test))
cvs_all.append(np.mean(scores))
print("Accuracy:{0:.2%}".format(accuracy_score(knn_y_pred, y_test)))
print("Cross Validation Score: {0: .2%}(+/- {1: .2%})".format(np.mean(scores), np.std(scores)*2))
```

```python
accuracy_all=[] #logistic regression
cvs_all=[]
scores=cross_val_score(knn, x, y, cv=10) #KNN
accuracy_all.append(accuracy_score(lr_y_pred, y_test))
cvs_all.append(np.mean(scores))
print("Accuracy:{0:.2%}".format(accuracy_score(lr_y_pred, y_test)))
print("Cross Validation Score: {0: .2%}(+/- {1: .2%})".format(np.mean(scores), np.std(scores)*2))
```
